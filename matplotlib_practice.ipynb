```python
%matplotlib inline
```

```python
# Import dependencies
import matplotlib.pyplot as plt
```

```python
# Set the x-axis to a list of strings for each month
x_axis = ["Jan", "Feb", "Mar", "April", "May", "June", "July", "Aug", "Sept", "Oct", "Nov", "Dec"]
```

```python
# Set the y-axis to a list of floats as sthe total fare in US dollars accumulated for each month
y_axis = [10.02, 23.24, 39.20, 35.42, 32.34, 27.04, 43.82, 10.56, 11.85, 27.90, 20.71, 20.09]
```

```python
# Create the plot
plt.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
fig = plt.figure()
ax = fig.add_subplot()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot with ax.plt()
ax = plt.axes()
ax.plot(x_axis, y_axis)
```

```python
# Create the plot and add a label for the legend
plt.plot(x_axis, y_axis, label = 'Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.plot(x_axis, y_axis, marker = "^", color = "green", linewidth = 2, label = 'Boston')

# Create labels for the x and y axes
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Set the y limit between 0 and 45
plt.ylim(0,45)

# Create a title
plt.title("PyBer Fare by Month")

# Add a grid
plt.grid()

# Add the legend
plt.legend()
```

```python
# Create a bar chart
plt.bar(x_axis, y_axis)
```

```python
# Create the plot
plt.bar(x_axis, y_axis, color = "green", label = 'Boston')

# Create the labels for the x and y axis
plt.xlabel("Date")
plt.ylabel("Fare ($)")

# Create a title
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot
plt.barh(x_axis, y_axis, color="magenta", label = 'Boston')

# Invert the months
plt.gca().invert_yaxis()

# Title the plot, x-axis, and y-axis
plt.ylabel("Date")
plt.xlabel("Fare ($)")
plt.title("PyBer Fare by Month")

# Add the legend
plt.legend()
```

```python
# Create the plot with ax.plt()
fig, ax = plt.subplots()
ax.bar(x_axis, y_axis, color = "black", label = 'Boston')
```

```python
# Create the plot with ax.plt()
fix, ax = plt.subplots()
ax.barh(x_axis, y_axis, color = "cyan", label = 'Chicago')

# Invert the y-axis
ax.invert_yaxis()

# Add a title and axes labels
ax.set_title("PyBer Fare by Month")
ax.set_xlabel("Fare ($)")
ax.set_ylabel("Date")

# Add a legend
ax.legend()
```

```python
fig, ax = plt.subplots()
ax.scatter(
    y_axis, x_axis, 
    c = "skyblue", alpha = 0.80,
    edgecolors = "black", linewidth = 2,
    s = [i * 5 for i in y_axis], 
    label = 'Boston')
ax.set_ylabel("Date")
ax.set_xlabel("Fare ($)")
ax.legend()
ax.invert_yaxis()
ax.set_xlim(0,50)
ax.set_title("PyBer Fare per Month")
```

```python
plt.pie(y_axis, labels = x_axis)
plt.show()
```

```python
plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
plt.pie(y_axis, 
        explode = explode_values,
        colors = colors,
        labels = x_axis, 
        autopct = '%.1f%%')
plt.show()
```

```python
fig, ax = plt.subplots(figsize = (8,8))
explode_values = (0, 0, 0.2, 0, 0, 0, 0.2, 0, 0, 0, 0, 0)
colors = ["slateblue", "magenta", "lightblue", "green", "yellowgreen", "greenyellow", "yellow", "orange", "gold", "indianred", "tomato", "mistyrose"]
ax.pie(y_axis, 
       explode = explode_values,
       labels = x_axis,
       colors = colors,
       startangle = 90,
       shadow = True,
       counterclock = False,
       autopct = '%.1f%%')
plt.show()
```
